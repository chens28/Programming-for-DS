### Problem 01 (15 points):

The data sets provided in the file "Class.csv" represents the number of points obtained by students at a quizz, where every row represents points obtained by a student.

Load the file and write a function which displays: 

* Number of students 

* Number of quizzes

* Average number of points for each student 

* Min number of points for each student 

* Max number of points for each student 

* Standard deviation of the number of points for each student 

* Average number of points for each quiz 

* Min number of points for each quiz 

* Max number of points for each quiz 

* Standard deviation of the number of points for each quiz 



````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
data = np.loadtxt(fname='Class.csv', delimiter=',')
data
````

````python
def summarize_data(array):
    print("Number of students:", array.shape[0])
    print("Number of quizzes:", array.shape[1])
    print("Average number of points for each student:", array.mean(axis = 1))
    print("Min number of points for each student:", array.min(axis = 1))
    print("Max number of points for each student:", array.max(axis = 1))
    print("Standard deviation of the number of points for each student:", array.std(axis = 1))
    print("Average number of points for each quiz:", array.mean(axis = 0))
    print("Min number of points for each quiz:", array.min(axis = 0))
    print("Max number of points for each quiz:", array.max(axis = 0))
    print("Standard deviation of the number of points for each quiz:", array.std(axis = 0).round(1))
````

````python
summarize_data(data)
````

3.2 (5 points) Apply the ```np.linalg.lstsq``` method to find the least-squares solution to the linear matrix equation `A.x = y`. Print the value of `x` found with this method to confirm that it is the same as the one found by explicit calculation in 3.1. 

3.3 (5 points) For the matrix `B` of new observations given below, use the value of the coefficient vector `x` found in 3.1 and 3.2 to predict the value of $\hat{y}$ according to the linear regression model $A.x = \hat{y}$



````python
import numpy as np

# Matrix A 
A = np.matrix([[1, 2], 
               [2, 3],
               [3, 4],
               [4, 5]])

# Matrix B
B = np.matrix([[1, 1], 
               [0, 2],
               [0, 1],
               [7, 2]])

# Vector y
y = np.matrix([[4], 
               [5],
               [8],
               [9]])
````

````python
#3.1 p1
AT = A.T
ATA = AT@A
np.linalg.det(ATA)
````

````python
#3.1 p2
ATAi = np.linalg.inv(ATA)
ATy = AT@y
x = ATAi@ATy
print(x)
````

````python
#3.2
results = np.linalg.lstsq(A, y, rcond = None)
print(results[0])
````

````python
#3.3
y_hat = B@x
y_hat
````

### Problem 04 (10 points)

4.1 (5 points) Plot in the same diagram the functions: $1, x, x^2, x^3, \dots, x^{10}$ over the following set of points $\{0, 0.01, 0.02, \dots, 1\}$, that is, over 101 equidistant points from the interval $[0, 1]$. 


4.2 (5 points) Plot in the same diagram the functions: $1, x, x^{1/2}, x^{1/3}, \dots, x^{1/10}$ over the following set of points $\{0, 0.01, 0.02, \dots, 1\}$, that is, over 101 equidistant points from the interval $[0, 1]$. 


````python
#4.1
x = np.linspace(0, 1, 101)

for i in range(11):
    plt.plot(x, x**i, label = f'x^{i}')

plt.xlabel('x')
plt.ylabel('x^i', rotation='horizontal')
plt.legend()
plt.show()
````

````python
#4.2
x = np.linspace(0, 1, 101)

plt.plot(x, x**0, label = f'x^{0.0}')
for i in range(1,11):
    plt.plot(x, x**(1/i), label = f'x^{round(1/i, 3)}')

plt.xlabel('x')
plt.ylabel('x^(1/i)', rotation='horizontal')
plt.legend()
plt.show()
````

* (m4, s4) = (1, 1) 

then plot each as a histogram with 50 bins on the same plot. Show a legend in the upper left corner of the plot with a custom label for each histogram (respectively `h1`, `h2`, `h3`, `h4`). Adjust the figure x-axis to go from $-7$ to $+7$, and the y-label to go from 0 to +80.


5.2 (15 points) 

Create a `Figure` object of width $10$ inches and height $8$ inches with $2 \times 2$ subplots (i.e. 2 subplots in each row and 2 in each column), and **plot each of the histograms above as a separate subplot** on this `Figure` object, using a tight layout.
````python
#5.1
n1 = np.random.normal(0, 2, 1000)
n2 = np.random.normal(2, 0.5, 1000)
n3 = np.random.normal(-2, 0.5, 1000)
n4 = np.random.normal(1, 1, 1000)

plt.hist(n1, bins = 50, label = 'h1', alpha = 0.8)
plt.hist(n2, bins = 50, label = 'h2', alpha = 0.8)
plt.hist(n3, bins = 50, label = 'h3', alpha = 0.8)
plt.hist(n4, bins = 50, label = 'h4', alpha = 0.8)
plt.legend(loc = 2)
plt.xlim(-7, 7)
plt.ylim(0, 80)
plt.show()
````

````python
#5.2
f, ax = plt.subplots(2, 2, figsize = (10, 8), tight_layout = True)

ax[0,0].hist(n1, bins = 50, label = 'h1', color = '#1f77b4')
ax[0,1].hist(n2, bins = 50, label = 'h2', color = '#ff7f0e')
ax[1,0].hist(n3, bins = 50, label = 'h3', color = '#2ca02c')
ax[1,1].hist(n4, bins = 50, label = 'h4', color = '#d62728')
plt.setp(ax, xlim = (-7, 7), ylim = (0, 80))
plt.show()
````
